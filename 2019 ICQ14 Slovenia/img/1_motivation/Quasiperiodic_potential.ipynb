```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

from sys import path
path.insert(0, "../../../../free-fermions/lib")
import geometry as geom
```

```python
slope = 0.5*(np.sqrt(5)-1)
seq = np.array(geom.words(slope, 50)[0])
```

```python
xi = np.arange(len(seq))
spl = UnivariateSpline(xi, seq)
spl.set_smoothing_factor(0.1)
x = np.linspace(10-0.5, 27+0.5, 1500)

fig, ax = plt.subplots()
plt.plot(x, spl(x), 'k')
ax.set_aspect(5)
plt.savefig('qp_curve.svg')
plt.show()
```
